### Position

```python
df2=pd.read_excel(r"C:\Users\gnana\Downloads\2023 Cricket dataset.xlsx",sheet_name="Batting_score_card_all_matches")
count=1
df2["Position"]=0
current_team = df2['innings'][0]
for i in range(0,len(df2)):
    if df2['innings'][i] != current_team:
        df2['Position'][i] = 1
        count = 2
        current_team = df2['innings'][i]
    else:
        df2['Position'][i] = count
        count+=1
df2
```

### Boundary percentage

```python
df2["Boundary"]=(df2["fours"]*4)+(df2["sixes"]*6)
ca=df2.groupby("Batsman")
df2["Runs"] = pd.to_numeric(df2["Runs"], errors='coerce')
runs=ca.Runs.sum()
df2["Boundary"] = pd.to_numeric(df2["Boundary"], errors='coerce')
boundaries=ca.Boundary.sum()
boundaryper=(boundaries/runs)*100 
bounds=pd.DataFrame(boundaryper)
bounds=bounds.reset_index().rename(columns={
    0:"BOundary per"
})
bounds.rename(columns={
    "Batsman":"Player",
    "BOundary per":"Boundary per"
},inplace=True)
bounds 
```

```python
topPlayer=df[df["Ave"]>50]
topPlayer=topPlayer[topPlayer["SR"]>=90]
topPlayer=topPlayer[topPlayer["Mat"]>=5]
topPlayer=topPlayer[topPlayer["Runs"]>=500]
topPlayer["Boundary %"]=np.NAN
topPlayer.drop(columns="Boundary %",inplace=True) 
topPlayer=topPlayer.merge(bounds,on="Player")
topPlayer=topPlayer.fillna(0)
topPlayer=topPlayer[topPlayer["Boundary per"]>40]
topPlayer
```

### Frequent position by a batsman

```python
result = df2.groupby("Batsman")["Position"].apply(lambda x: x.mode().iloc[0]).reset_index()
```

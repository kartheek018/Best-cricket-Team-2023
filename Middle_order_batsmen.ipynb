```python
df["Inns"].replace('-',0,inplace=True)
```

### Innings Average Strike rate

```python
Middle_players=df[df["Inns"]>8]
Middle_players=Middle_players[Middle_players["Ave"]>50]
Middle_players=Middle_players[Middle_players["SR"]>100]
```

### Balls faced By each player overall Tournment

```python
df2["Balls"].replace('-',0,inplace=True)
Balls_faced=df2.groupby("Batsman")["Balls"].apply(lambda x:x.sum())
Balls_faced=pd.DataFrame(Balls_faced)
Balls_faced.reset_index(inplace=True)
Balls_faced.rename(columns={
    "Batsman":"Player"
},inplace=True)
if 'Balls' not in Middle_players.columns:
    Middle_players = Middle_players.merge(Balls_faced, on="Player")
else:
    # Drop the existing column before merging the updated one
    Middle_players.drop(columns=['Balls'], inplace=True)
    Middle_players = Middle_players.merge(Balls_faced, on="Player")

Middle_players
```

### Balls Faced

```python
Middle_players=Middle_players[Middle_players["Balls"]>100]
Middle_players
```

### Position of a player

```python
count=1
df2["Position"]=0
current_team = df2['innings'][0]
for i in range(0,len(df2)):
    if df2['innings'][i] != current_team:
        df2['Position'][i] = 1
        count = 2
        current_team = df2['innings'][i]
    else:
        df2['Position'][i] = count
        count+=1
result = df2.groupby("Batsman")["Position"].apply(lambda x: x.mode().iloc[0]).reset_index()
result.rename(columns={
    "Batsman":"Player"
},inplace=True)
```

```python
Middle_players=Middle_players.merge(result,on="Player")
Middle_players=Middle_players[(Middle_players["Position"]>3) & (Middle_players["Position"]<6)]
Middle_players 
```

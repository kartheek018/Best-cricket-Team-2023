```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\gnana\Downloads\2023 Cricket dataset.xlsx",sheet_name="Bowling_data")
df.replace('-',0,inplace=True)
df.head()
```

```python
df2=pd.read_excel(r"C:\Users\gnana\Downloads\2023 Cricket dataset.xlsx",sheet_name="Batting_score_card_all_matches")
count=1
df2["Position"]=0
current_team = df2['innings'][0]
for i in range(0,len(df2)):
    if df2['innings'][i] != current_team:
        df2['Position'][i] = 1
        count = 2
        current_team = df2['innings'][i]
    else:
        df2['Position'][i] = count
        count+=1
result = df2.groupby("Batsman")["Position"].apply(lambda x: x.mode().iloc[0]).reset_index()
result.rename(columns={
    "Batsman":"Player"
},inplace=True)
```

```python
df=df.merge(result,on="Player")
```

```python
Bowlers=df[df["Position"]>=8]
```
